```python
x=np.arange(0,10)
y= 3*x + 10
```

```python
plt.plot(x, y, 'go--')
plt.plot(x, y, 'r*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X-Axis' , color='blue')
plt.ylabel('Y-Axis', color='blue')
plt.title('2d diagram', color='blue')
```

Creating sub plots 

```python
plt.subplot(2,2,1)
plt.plot(x,y, 'r--')
plt.subplot(2,2,2)
plt.plot(x,y, 'g*--')
plt.subplot(2,2,3)
plt.plot(x,y, 'bo')
plt.subplot(2,2,4)
plt.plot(x,y, 'go')
```
