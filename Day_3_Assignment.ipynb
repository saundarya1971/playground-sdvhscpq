```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri =sns.load_dataset('fmri')
fmri.head()
```

```python
sns.scatterplot(x='subject', y='timepoint', data = fmri, hue='region')
```

```python
sns.relplot(x='subject',y='timepoint',data=fmri,style='region',size='signal')
```

```python
sns.catplot(x='region',y='timepoint',data=fmri)
```

```python
sns.catplot(x='event',y='timepoint',data=fmri,hue='region',kind='swarm')
```

```python
sns.catplot(x='event',y='timepoint',data=fmri, kind='violin',hue='region',style='subject')
```
